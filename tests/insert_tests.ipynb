```python
quarters = ['2019-q1', '2019-q2', '2019-q3', '2019-q4',
            '2020-q1', '2020-q2', '2020-q3', '2020-q4',
            '2021-q1', '2021-q2', '2021-q3', '2021-q4']

total_time = 0
```

```python
column_names = ["year", "quarter", "quadkey", "avg_d_kbps", "avg_u_kbps",
                 "avg_lat_ms", "tests", "devices"]

for quarter in quarters:
    year = int(quarter.split("-")[0])
    quarter_num = int(quarter.split("-")[1][-1])
    
    data = pd.read_parquet(os.path.join("tests","data", "internet_speed_dataset",
                                    quarter, "2019-01-01_performance_fixed_tiles.parquet"))
    
    for row in data.iterrows():
        values = [year, quarter_num, row[1]['quadkey'], row[1]['avg_d_kbps'], 
                    row[1]['avg_u_kbps'], row[1]['avg_lat_ms'], row[1]['tests'], 
                    row[1]['devices']]
        
        sql_executer.insert(table_name = 'internet_data',
                            columns = column_names,
                            values = values)
        
    data = pd.read_parquet(os.path.join("tests","data", "internet_speed_dataset",
                                    quarter, "2019-01-01_performance_mobile_tiles.parquet"))
    
    for row in data.iterrows():
        values = [year, quarter_num, row[1]['quadkey'], row[1]['avg_d_kbps'], 
                    row[1]['avg_u_kbps'], row[1]['avg_lat_ms'], row[1]['tests'], 
                    row[1]['devices']]
        
        sql_executer.insert(table_name = 'internet_data',
                            columns = column_names,
                            values = values)
```
